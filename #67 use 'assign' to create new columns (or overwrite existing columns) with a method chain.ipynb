```python
import pandas as pd 
```

```python
drinks = pd.read_csv('https://raw.githubusercontent.com/justmarkham/DAT8/master/data/drinks.csv')
drinks.head()
```

```python
drinks = drinks.set_index('country')
drinks.head()
```

```python
df = drinks.loc[:,['continent','beer_servings']]
df.head()
```

```python
(df.assign(continent = df.continent.str.lower(),
          beer_ounce = df.beer_servings * 12,
          beer_gallons = lambda df: df.beer_ounce / 128)
.query('beer_gallons > 30')
.style.set_caption('Average beer consumption per person in 2010'))
```
